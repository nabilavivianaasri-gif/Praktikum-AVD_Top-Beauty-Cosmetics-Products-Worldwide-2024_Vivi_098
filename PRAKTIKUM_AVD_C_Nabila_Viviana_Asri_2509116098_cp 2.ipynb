## **Duplicated Values**

```python
filenyanih[filenyanih.duplicated()]
```

**Kesimpulan:** Tidak ada Duplicated Values pada dataset ini.

```python
results = []

cols = filenyanih.select_dtypes(include=['float64', 'int64'])

for col in cols:
    q1 = filenyanih[col].quantile(0.25)
    q3 = filenyanih[col].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = filenyanih[(filenyanih[col] < lower_bound) |
                           (filenyanih[col] > upper_bound)]

    percent_outliers = (len(outliers) / len(filenyanih)) * 100

    results.append({'Kolom': col,'Persentase Outliers (%)': percent_outliers})

results_filenyanih = pd.DataFrame(results)
results_filenyanih.set_index('Kolom', inplace=True)

display(results_filenyanih)
```

## **Outliers**

```python
plt.subplot(1, 3, 1)
sns.boxplot(y=filenyanih['Price_USD'])
plt.title('Boxplot of Price (USD)')

plt.subplot(1, 3, 2)
sns.boxplot(y=filenyanih['Rating'])
plt.title('Boxplot of Rating')

plt.subplot(1, 3, 3)
sns.boxplot(y=filenyanih['Number_of_Reviews'])
plt.title('Boxplot of Number of Reviews')

plt.tight_layout()
plt.show()
```

**Kesimpulan:** Terdapat beberapa outliers pada Price_USD, Rating, dan Number_of_Reviews berdasarkan metode IQR. Namun, jika persentasenya relatif kecil dibandingkan total 15.000 data, maka outliers tersebut tidak terlalu signifikan dan masih dapat ditoleransi dalam analisis.

Outliers pada Price_USD kemungkinan berasal dari produk dengan harga sangat rendah atau sangat tinggi dibandingkan mayoritas produk. Pada Rating, outliers bisa muncul karena adanya produk dengan penilaian ekstrem (sangat rendah atau sangat tinggi). Sedangkan pada Number_of_Reviews, outliers biasanya berasal dari produk yang sangat populer dengan jumlah ulasan jauh di atas rata-rata.

Secara keseluruhan, selama persentase outliers kecil, kondisi ini tidak terlalu mengganggu hasil analisis dan justru dapat memberikan insight tambahan mengenai produk-produk yang memiliki karakteristik ekstrem di dalam dataset.


# **Eksplorasi Data (EDA)**

## **Comparison(Perbandingan)**

```python
avg_rating_by_category = filenyanih.groupby('Category')['Rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=avg_rating_by_category.index, y=avg_rating_by_category.values, palette='viridis', hue=avg_rating_by_category.index, legend=False)
plt.title('Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

**Kesimpulan:** Rata-rata rating pada seluruh kategori produk berada di kisaran yang hampir sama, yaitu sekitar 2.9 hingga 3.1. Tidak terdapat perbedaan yang signifikan antar kategori, sehingga dapat disimpulkan bahwa tingkat kepuasan pelanggan relatif merata di semua jenis produk.

Kategori seperti Contour dan Moisturizer sedikit lebih tinggi dibandingkan kategori lainnya, namun selisihnya sangat kecil sehingga secara umum performa rating antar kategori tergolong stabil dan konsisten.

## **Composition(Komposisi)**

```python
gender_counts = filenyanih['Main_Ingredient'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Proporsi Produk Berdasarkan Bahan Utama')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

**Kesimpulan:** Berdasarkan diagram pie tersebut, komposisi Main_Ingredient menunjukkan bahwa seluruh bahan aktif memiliki persentase yang hampir seimbang, yaitu di kisaran 13%–15%. Tidak ada satu bahan yang mendominasi secara signifikan.

Hal ini menunjukkan bahwa distribusi kandungan utama produk dalam dataset cukup merata, sehingga variasi produk lebih tersebar dan tidak terfokus pada satu jenis bahan tertentu saja.


## **Distribution(Distribusi)**

```python
plt.figure(figsize=(10, 6))
sns.histplot(filenyanih['Rating'], bins=10, kde=True, color='skyblue')
plt.title('Distribusi Rating Produk')
plt.xlabel('Rating')
plt.ylabel('Frekuensi')
plt.show()
```

**Kesimpulan:** Distribusi rating produk terlihat cukup merata dari nilai 1 hingga 5. Tidak ada penumpukan yang terlalu ekstrem pada satu nilai tertentu, meskipun rating di sekitar angka 3–4 tampak sedikit lebih dominan.

Hal ini menunjukkan bahwa sebagian besar produk memiliki penilaian yang cukup baik (kategori sedang hingga tinggi), dan tidak terdapat distribusi yang terlalu condong ke rating sangat rendah maupun sangat tinggi.


## **Relationship/Hubungan**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=filenyanih[['Rating', 'Number_of_Reviews']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Rating dan Jumlah Ulasan')
plt.show()
```

**Kesimpulan:** Berdasarkan heatmap korelasi, nilai korelasi antara Rating dan Number_of_Reviews adalah sekitar -0.00 atau mendekati 0.

Hal ini menunjukkan bahwa tidak terdapat hubungan linear yang signifikan antara rating produk dan jumlah ulasan. Artinya, banyaknya jumlah review tidak secara langsung memengaruhi tinggi atau rendahnya rating produk dalam dataset ini.
